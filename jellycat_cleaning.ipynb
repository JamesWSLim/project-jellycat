```python
import pandas as pd
import numpy as np

# import data
df = pd.read_csv("jellycat.csv")

# remove repititive index column
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df
```

```python
# extract size from information
df[['size', 'information']] = df['information'].str.split('Buy MeOut of Stock', n=1, expand=True)
df[['size', 'useless']] = df['size'].str.split(' -', n=1, expand=True)
df
```

```python
df[['information', 'price']] = df['information'].str.split('$', n=1, expand=True)
df[['price', 'useless']] = df["price"].str.split(" ", n=1, expand=True)
df
```

```python
df["stock availability"] = "In Stock"
df.loc[df['information'].str.contains("Out of Stock"), "stock availability"] = "Out of Stock"
df.loc[df['information'].str.contains("Available in"), "stock availability"] = "Out of Stock"
df.loc[df['information'].str.contains("Back in stock"), "stock availability"] = "Out of Stock"
```
